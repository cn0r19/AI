```python
def IsGoalReachable(graph, startNode, goalNode, visited=None):

    if visited is None:
        visited = set()


    if startNode == goalNode:
         return 'SUCCESS'

    visited.add(startNode)

    # Recursively visit unvisited neighbors
    for neighbor in graph.get(startNode, []):
        if neighbor not in visited:
          result = IsGoalReachable(graph, neighbor, goalNode, visited)
          if result == 'SUCCESS':
            return 'SUCCESS'

    return 'FAILURE'
```
