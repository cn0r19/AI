```python

class WaterJugPuzzle():

  def __init__(self): #this project uses the lists big and small to represent the two jugs, both will start empty
    self.big = list()
    self.small = list()

  def is_big_below_0(self):
    if len(self.big) < 0:
      return True

  def is_small_below_0(self):
    if len(self.small) < 0:
      return True

  def is_big_empty(self):
    if not self.big:
      return True

  def is_small_empty(self):
    if not self.small:
      return True

  def is_big_jug_overfilled(self):
    if len(self.big) > 4:
      return True

  def is_small_jug_overfilled(self):
    if len(self.small) > 2:
      return True

  def state_ok(self):
    return not (self.is_big_jug_overfilled() or self.is_small_jug_overfilled() or self.is_big_below_0() or self.is_small_below_0())

  def save_state(self):
    self._big = self.big
    self._small = self.small

  def undo_state(self):
    self.big = self._big
    self.small = self._small

  ###############################################################################
  # Rule 1: Fill big jug full from environment
  ###############################################################################

  def R1_fill_big_from_outside(self):

    self.save_state()

    for i in range(5):
      if not self.is_big_jug_overfilled():
        self.big.append('water')

    if not self.state_ok(): self.undo_state()


  #################################################################################
  # Rule 2: fill small jug full from environment
  #################################################################################

  def R2_fill_small_from_outside(self):

    self.save_state()

    for i in range(3):
      if not self.is_small_jug_overfilled():
        self.small.append('water')

    if not self.state_ok(): self.undo_state()

  #################################################################################
  # Rule 3: Fill small jug from big jug
  #################################################################################

  def R3_add_to_small_from_big(self):

    self.save_state()

    while not (self.is_small_jug_overfilled() or self.is_big_empty()):
      if 'water' in self.big:
        self.big.remove('water')
        self.small.append('water')
        if not self.state_ok(): self.undo_state()
      else:
        raise ValueError('the large jug cannot move any water to the small jug')

  #################################################################################
  # Rule 4: Fill big jug from small jug
  #################################################################################

  def R4_add_to_big_from_small(self):

    self.save_state()

    while not (self.is_big_jug_overfilled() or self.is_small_empty()):
      if 'water' in self.small:
        self.small.remove('water')
        self.big.append('water')
        if not self.state_ok(): self.undo_state()
      else:
        raise ValueError('the small jug is empty, no water to be moved')

  #################################################################################
  # Rule 5: Empty big jug to environment
  #################################################################################

  def R5_pour_out_big(self):

    self.save_state()

    while not self.is_big_empty():
      if 'water' in self.big:
        self.big.remove('water')
        if not self.state_ok(): self.undo_state()
      else:
        raise ValueError('the big jug was empty already')

  #################################################################################
  # Rule 6: Empty small jug to environment
  #################################################################################

  def R6_pour_out_small(self):

    self.save_state()

    while not self.is_small_empty():
      if 'water' in self.small:
        self.small.remove('water')
        if not self.state_ok(): self.undo_state()
      else:
        raise ValueError('the small jug was empty already')

  #################################################################################
  # Invariants
  #################################################################################

  def big_jug_is_legal(self):
    return not (self.is_big_below_0() or self.is_big_jug_overfilled())

  def small_jug_is_legal(self):
    return not (self.is_small_below_0() or self.is_big_small_overfilled())

  #################################################################################
  # Print State of system
  #################################################################################

  def print_state(self):
    print("===> big: {s}, small: {h}".format(s=self.big, h=self.small))
```

```python
#@title Solution

logic = WaterJugPuzzle()

logic.print_state()

print('---------------------------')
logic.R1_fill_big_from_outside()
logic.print_state()

print('---------------------------')
logic.R3_add_to_small_from_big()
logic.print_state()

print('---------------------------')
logic.R6_pour_out_small()
logic.print_state()

print('---------------------------')
logic.R3_add_to_small_from_big()
logic.print_state()

print('---------------------------')
logic.R1_fill_big_from_outside()
logic.print_state()

print('---------------------------')
logic.R3_add_to_small_from_big()
logic.print_state()

```
